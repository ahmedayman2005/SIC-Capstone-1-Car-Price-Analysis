###6.What is the highest average MRSP Makes and it's most expensive model ?


```python
avg_price_by_make = (df.groupby('Make')['MSRP'].mean().sort_values(ascending=False).head(15))
most_expensive_models = df.loc[df.groupby('Make')['MSRP'].idxmax(),['Make', 'Model', 'MSRP']]
top_makes = (avg_price_by_make.reset_index().merge(most_expensive_models, on='Make', how='left'))
top_makes = top_makes.rename(
    columns={
        'MSRP_x': 'Average MSRP',
        'MSRP_y': 'Most Expensive MSRP'
    }
)
top_makes['Car Label'] = top_makes['Make'] + ' ' + top_makes['Model']

plt.figure(figsize=(12, 7))
sns.barplot(x='Average MSRP',y='Make',data=top_makes,palette='viridis')
plt.scatter(top_makes['Most Expensive MSRP'],top_makes['Make'],color='purple',s=50,marker='D',edgecolor='black',linewidth=0.4,zorder=5,label='Most Expensive Model')

for price, label, make in zip(top_makes['Most Expensive MSRP'],top_makes['Car Label'],top_makes['Make']):
    plt.text(price + 800, make, label, va='center', fontsize=9, color='purple', fontweight='bold')

plt.title('Average MSRP per Make with Most Expensive Model (Top 15)',fontsize=16,fontweight='bold')
plt.xlabel('Price (MSRP)')
plt.ylabel('Car Make')
plt.legend()
plt.tight_layout()
```

####Observation:
    1.Bugatti has both the highest average MSRP and the most expensive model overall (Bugatti Veyron 16.4 at over $2M).
    2.For most luxury brands, the most expensive model is far above the brands average price, especially forniche luxury brands like Maybach, Rolls-Royce and Lamborghini
    3.There is a clear gap between the average car price and the flagship models showing how a single ultra-luxury flagship can influence brand perception.

#6.Feature Engineering

```python
df_scaled = pd.read_csv("my_data.csv")

df_scaled.describe().T

df_scaled.columns
```

```python
#Car Age
df_scaled['Age'] = 2017 - df_scaled['Year']

#HP per Cylinder, completely ignoring the electric cars
df_scaled['hp_per_cylinder'] = df_scaled['Engine HP'] / df_scaled['Engine Cylinders'].replace(0, np.nan)


#MSRP per HP (price efficiency)
df_scaled['msrp_per_hp'] = df_scaled['MSRP'] / df_scaled['Engine HP']
#Avoid division by zero if Engine HP has 0 which is unlikely now
df_scaled['msrp_per_hp'].replace([np.inf, -np.inf], np.nan, inplace=True)
df_scaled['msrp_per_hp'].fillna(df_scaled['MSRP'], inplace=True)

print(df_scaled[['Age','hp_per_cylinder','msrp_per_hp','has_market_category']].head())
```

```python
#Remove EVs before plotting
df_non_ev = df_scaled[
    ~df_scaled["Engine Fuel Type"].str.contains("electric", case=False, na=False)
]  #only keep petrol cars

plt.figure(figsize=(10,6))
sns.scatterplot(
    x='hp_per_cylinder',
    y='msrp_per_hp',
    data=df_non_ev,
    hue='Vehicle Size_cat',
    alpha=0.7
)
plt.title('MSRP per HP vs HP per Cylinder (Non-EVs)')
plt.xlabel('HP per Cylinder')  #Engine efficiency
plt.ylabel('MSRP per HP ($)')  #Cost efficiency
plt.legend(title='Vehicle Size')
plt.show()


```

###Observations:
    1.The plot shows a clear spread of cars across different vehicle sizes.
    2.Compact and midsize cars cluster around lower hp_per_cylinder and lower msrp_per_hp, indicating balanced efficiency and cost.
    3.Large vehicles generally appear at higher msrp_per_hp, showing that they tend to be less cost-efficient per horsepower.
    4.There are some outliers with very high msrp_per_hp despite moderate hp_per_cylinder, suggesting overpriced models.

###Summary
    1.Overall, compact and midsize cars provide better value for money in terms of cost per horsepower.
    2.Large vehicles, while more powerful, are less efficient and more expensive relative to the performance they offer.
    3.The presence of outliers highlights models that deviate strongly from typical cost-performance trends.

```python
#No filtering by age or MSRP!

#Pick top/bottom 3 cars by value-for-money
best_value = df_non_ev.nsmallest(3, "msrp_per_hp")   # lowest = best value
worst_value = df_non_ev.nlargest(3, "msrp_per_hp")   # highest = worst value

def make_label(row, df):
    #if this (Make, Model) appears multiple times in comparison set, print car year as well
    subset = df[df["Make"] == row["Make"]]
    if subset["Model"].nunique() == 1 and subset.shape[0] > 1:
        return f"{row['Make']} {row['Model']} ({row['Year']})"
    else:
        return f"{row['Make']} {row['Model']}"

best_value = best_value.copy()
worst_value = worst_value.copy()

best_value["label"] = best_value.apply(lambda r: make_label(r, best_value), axis=1)
worst_value["label"] = worst_value.apply(lambda r: make_label(r, worst_value), axis=1)

comparison = pd.concat([best_value, worst_value])
comparison = comparison.sort_values("msrp_per_hp")

plt.figure(figsize=(12,6))

plt.stem(comparison["label"], comparison["msrp_per_hp"],
         linefmt="gray", markerfmt="o", basefmt=" ")

#best (green) and worst (red)
plt.scatter(best_value["label"], best_value["msrp_per_hp"],
            color="green", s=100, label="Best Value")
plt.scatter(worst_value["label"], worst_value["msrp_per_hp"],
            color="red", s=100, label="Worst Value")

plt.xticks(rotation=45, ha="right")
plt.title("Value for Money Ranking (MSRP per HP)")
plt.ylabel("MSRP per HP ($)")
plt.legend()
plt.tight_layout()
plt.show()
```

####Observation:
    1.The three best value-for-money cars are from the early 1990s to mid-1990s, they are: Lincoln Mark VIII (1996), Audi V8 (1992),
    and Mercedes-Benz 600-Class (1992).
    2.These cars deliver good horsepower at an extremely low MSRP per horsepower (around 7 dollars per HP),so they standouts in cost efficiency.
    3.However, the three worst value-for-money cars are ultra-luxury models: Lamborghini Reventon (2008) and two versions of the Maybach Landaulet (2011, 2012).
    4.These cars have MSRP per horsepower above 2200 dollars, reflecting how luxury branding inflate cost far beyond performance value.
    5.Overall, older luxury sedans from the 1990s appear to be exceptional in terms of performance per dollar, though their age and practicality today may limit real-world appeal.
    6.In contrast, modern ultra-luxury brands such as Lamborghini and Maybach charge extremely high prices that are not justified by performance alone.

```python
#Filter by MSRP AND Age
#average car buyer price range ("People's car") AND age filter
lower_limit = 5000
upper_limit = 40000

df_people = df_non_ev[
    (df_non_ev["MSRP"] >= lower_limit) &
    (df_non_ev["MSRP"] <= upper_limit) &
    (df_non_ev["Age"] <= 10)   # only cars 10+ years old
]

best_valueForMoney = df_people.nsmallest(3, "msrp_per_hp")
worst_valueForMoney = df_people.nlargest(3, "msrp_per_hp")

def make_label(row, df):
    subset = df[(df["Make"] == row["Make"]) & (df["Model"] == row["Model"])]
    if subset.shape[0] > 1:
        return f"{row['Make']} {row['Model']} ({row['Year']})"
    else:
        return f"{row['Make']} {row['Model']}"

best_valueForMoney["label"] = best_valueForMoney.apply(lambda r: make_label(r, df_people), axis=1)
worst_valueForMoney["label"] = worst_valueForMoney.apply(lambda r: make_label(r, df_people), axis=1)


comparison_people = pd.concat([best_valueForMoney, worst_valueForMoney])
comparison_people = comparison_people.sort_values("msrp_per_hp")


plt.figure(figsize=(12,6))

plt.stem(comparison_people["label"], comparison_people["msrp_per_hp"],
         linefmt="gray", markerfmt="o", basefmt=" ")

plt.scatter(best_valueForMoney["label"], best_valueForMoney["msrp_per_hp"],
            color="green", s=100, label="Best Value")
plt.scatter(worst_valueForMoney["label"], worst_valueForMoney["msrp_per_hp"],
            color="red", s=100,
             label="Worst Value")


plt.xticks(rotation=45, ha="right")
plt.title("People's Car Value-for-Money Ranking (MSRP per HP, $5k$40k, Age 10 yrs)")
plt.ylabel("MSRP per HP ($)")
plt.legend()
plt.tight_layout()
plt.show()
```

####Observations:
    1.The Chevrolet Camaro (2015) is the best value-for-money option, offering a strong horsepower output at a competitive MSRP.
    2.Its msrp_per_hp of around 73 indicates good cost efficiency compared to other cars in the same affordability and age range.
    3.But, Acura ILX Hybrid is the worst, with a msrp_per_hp above 310.
    4.Thus we notice limited performance of the Acura ILX Hybrid despite its price point
    5.The analysis shows the difference, a performance-oriented car like the Chevrolet Camaro delivers more horsepower per dollar, and this appeals to buyers prioritizing raw value.
    6.Meanwhile, hybrid options such as the Acura ILX Hybrid reflect compromises made for efficiency and eco-friendliness, leading to weaker cost efficiency when judged by power output.
